```python
import pandas as pd

data = {'дата': [], 'врач': [], 'оценка': [], 'комментарий': []}
with open('/content/drive/MyDrive/ML_learn/отзывы.txt', 'r') as f:
  f_data = f.readlines()
  for _ in f_data:
    _ = _.strip()
    name_col = _[:_.find(':')].lower()
    value = _[_.find(':') + 2:].strip()
    if name_col.startswith('1') == True:
      name_col = name_col[name_col.find('оценка'):].strip()
    if name_col != '' and value != '':
      if value.isdigit() == True:
        value = int(value)
      data[name_col].append(value)


df = pd.DataFrame(data = data)
df.to_csv('/content/drive/MyDrive/ML_learn/comments.csv', index = False)
```
